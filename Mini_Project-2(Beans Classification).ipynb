```python
# Applying label encoding of target column
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Class'] = le.fit_transform(df['Class'])
```

```python
sns.countplot(x='Class',data=df)
plt.show()
```

## Splitting Feature & Target

```python
X = df.drop(['Class'],axis=1)
y = df['Class']
```

```python
# Feature 

X
```

```python
# Target

y
```

## Splitting data into Training & Tetsing Data

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
# Printing shape

print("X_train :",X_train.shape)
print("X_test :",X_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

## Model Training

```python
# Loading the SVM Classification model

from sklearn.svm import SVC

clf = SVC(kernel = 'rbf',random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.fit_transform(X_test)
```

## Fitting Model

```python
clf.fit(X_train_scaled,y_train)
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
# Checking Accuracy on Training Data

acc1 = clf.score(X_train_scaled,y_train)
print("Accuracy on Training data :",acc1)
```

```python
# Prediction on Testing Data

y_pred = clf.predict(X_test_scaled)
print(y_pred)
```

```python
# Checking Accuracy on Testing Data

acc2 = accuracy_score(y_pred,y_test)
print("Accuracy on Tetsing data :",acc2)
```

```python
# Confusion Matrix

cm = confusion_matrix(y_pred,y_test)
print('Confusiom Matrix :')
print(cm)
```
